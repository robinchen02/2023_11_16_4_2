```python
import pandas as pd

scores = [
    {'chinese':78,'english':80,'math':92,'history':75,'discover':85},
    {'chinese':92,'english':60,'math':85,'history':62,'discover':73}
]
dataFrame = pd.DataFrame(index=["徐國堂","徐家家"],data=scores)
display(dataFrame.T)
```

```python
subjects = list(dataFrame.columns)
subjects
```

```python
dataFrame['chinese']
```

```python
swapDataFrame = dataFrame.T
display(swapDataFrame)
robert_series = swapDataFrame['robert']
display(robert_series)
jenny_series = swapDataFrame['jenny']
display(jenny_series)
```

```python
robert_scores = robert_series.to_list()
jenny_scores = jenny_series.to_list()
display(robert_scores)
display(jenny_scores)
```

```python
import matplotlib.pyplot as plt
figure = plt.figure()
axes = figure.add_subplot(1,1,1)
subjects = list(dataFrame.columns)
axes.plot(subjects,robert_scores,'bo-',label ='robert' )
axes.plot(subjects,jenny_scores,'ro-',label ='jenny')

axes.legend()
plt.show()
```
